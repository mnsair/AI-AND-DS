```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib_inline
```

```python
# Import titanic data from seaborn library
df = sns.load_dataset('titanic')
```

```python
# Create a plot
df.plot(kind='scatter', x = 'age', y = 'fare')
```

```python
# Change plot color
df.plot(kind='scatter', x = 'age', y = 'fare', color = 'red')
```

```python
# Change color by hex
# go to google and serch 'Color Picker' to select the color.
df.plot(kind='scatter', x = 'age', y = 'fare', color = '#AB7A3D')

# One can also use color picker chroem extention to pick any color, and have its HEX code here.
```

```python
# Group by plot
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
# Sex and Age plot with color
df.groupby('sex')['age'].mean().plot(kind='bar', color=['pink', 'blue'])
```

```python
# Plot with labels
df.groupby('sex')['age'].mean().plot(kind='bar', color=['pink', 'blue'])
plt.ylabel('AGE')
plt.xlabel('Gender')
```

```python
# seaborn plots
# bar plots using seaborn on df

sns.barplot(data=df, x = 'class', y = 'fare', hue='sex')
#hue used to further categorize or group the data based on 3rd variable, here we use `sex`. heu is seaborn command.
```

```python
#boxplot for the above data
sns.boxplot(data=df, x='sex', y = 'age', hue = 'class')
```

```python
# Scatter Plot only for numeric values comparison
sns.scatterplot(data = df, x = 'fare', y = 'age', hue = 'class')
```

```python
#who paid the highest fare
sns.scatterplot(data = df, x = 'fare', y = 'age', hue = 'sex')

# its male
```

```python
# shift the plot by changing the axies
sns.scatterplot(data = df, y = 'fare', x = 'age', hue = 'sex')
```
