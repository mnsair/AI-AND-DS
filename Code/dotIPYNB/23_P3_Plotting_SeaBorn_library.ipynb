# Seaborn and Plotting


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import things for Plotting according to grammar of Graphis
1. Data
2. Type of Plot
3. x axis
4. y axis
5. grouping (hue/color) base on categorical variable
6. size of the plot
7. color palletes of the plot
8. style of the plot
9. how do you interpret

```python
# Data
df = sns.load_dataset('titanic')
df.info()
```

```python
# Scatter Plot
# Scallter plot is for 2 numeric values.
sns.scatterplot(x='age', y='fare', data=df, hue = 'sex')
```

```python
# Scatter Plot
# Scallter plot is for 2 numeric values.
# Increase the size of the plot
# Estimated size for A$ page = 10:8 and slide = 16:9
plt.figure(figsize=(16,9))
sns.scatterplot(x='age', y='fare', data=df, hue = 'sex')
```

```python
# Scatter Plot
# Scallter plot is for 2 numeric values.
# Increase the size of the plot
# Estimated size for A$ page = 10:8 and slide = 16:9
# change the scatter point colors
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='fare', data=df, hue='sex', palette=['green', 'pink'])
```

```python
# Label the plot
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Scatter Plot of Age vs Fare")

#save the plot
plt.savefig(r'C:\Users\mnsai\Desktop\Python\AI-and-DS\Output\Plots\scatter_titanic.png', dpi=300)
plt.savefig(r'C:\Users\mnsai\Desktop\Python\AI-and-DS\Output\Plots\scatter_titanic.svg') #good to zoom-in as high as needed.
```

```python
# stle of a plot
# https://seaborn.pydata.org/generated/seaborn.set_style.html
sns.set_style("whitegrid")
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Scatter Plot of Age vs Fare")
```

```python
sns.set_theme(font_scale=1.5, font= 'Times New Roman') # set_theme() it self will also work, one can set font, stype, and other things inside the ()
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Scatter Plot of Age vs Fare")
```

# Box Plot

```python
sns.set_theme(font_scale=1.5, font= 'Times New Roman') # set_theme() it self will also work, one can set font, stype, and other things inside the ()
plt.figure(figsize=(10,8))
sns.boxplot(x='sex', y='age', data=df, hue='sex',palette=['green', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Box Plot of Gender vs Age")
```

```python
# find class unique values
df['class'].unique()
```

```python
sns.set_theme(font_scale=1.5, font= 'Times New Roman') # set_theme() it self will also work, one can set font, stype, and other things inside the ()
plt.figure(figsize=(10,8))
# you can put colors in palette or you can caeate palette as below and set palette = palette
# palette = {'First':'green', 'Second':'blue','Third': 'red'}
sns.boxplot(x='sex', y='age', data=df, hue='class',palette=['#550300', '#00dfaf', '#B0D1ED'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Box Plot of Gender, Age, and Class")

# chagne legend position and title
plt.legend(loc='upper right', title= 'Class Name')
```

## Boxen Plot

```python
sns.set_theme(font_scale=1.5, font= 'Times New Roman') # set_theme() it self will also work, one can set font, stype, and other things inside the ()
plt.figure(figsize=(10,8))
# you can put colors in palette or you can caeate palette as below and set palette = palette
palette = {'First':'green', 'Second':'blue','Third': 'red'}
sns.boxenplot(x='sex', y='age', data=df, hue='class',palette=palette)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Boxen Plot of Gender, Age and Class")

# chagne legend position and title
plt.legend(loc='upper right', title= 'Class Name')
```

```python
sns.set_theme(style='whitegrid', font_scale=1.5, font= 'Times New Roman') # set_theme() it self will also work, one can set font, stype, and other things inside the ()
plt.figure(figsize=(14,8))
# you can put colors in palette or you can caeate palette as below and set palette = palette
palette = {'First':'green', 'Second':'blue','Third': 'red'}
sns.boxenplot(x='sex', y='age', data=df, hue='class',palette=palette)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Boxen Plot of Gender, Age and Class")

# chagne legend position and title
plt.legend(loc='upper right', title= 'Class Name')

#change the x-ticks labels
# plt.xticks(rotation=45)  # Rotates ticks by 45 degrees
plt.xticks(ticks=[0,1], labels=['Male', 'Female'],rotation=45)

plt.savefig(r'C:\Users\mnsai\Desktop\Python\AI-and-DS\Output\Plots\boxed_titanic.svg') #SVG good to zoom-in as high as needed.
```

# Bar Plot using seaborn

```python
sns.set_theme(style='whitegrid', font_scale=1.5, font= 'Times New Roman') # set_theme() it self will also work, one can set font, stype, and other things inside the ()
plt.figure(figsize=(14,8))
# you can put colors in palette or you can caeate palette as below and set palette = palette
palette = {'First':'green', 'Second':'blue','Third': 'red'}
sns.barplot(x='sex', y='age', data=df, hue='class',palette=palette, capsize=0.1)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Boxen Plot of Gender, Age and Class")

# chagne legend position and title
plt.legend(loc='upper right', title= 'Class Name')

#change the x-ticks labels
# plt.xticks(rotation=45)  # Rotates ticks by 45 degrees
plt.xticks(ticks=[0,1], labels=['Male', 'Female'],rotation=45)

# # add caps to the error bars
# plt.errorbar(x=[0,1], y=[25,30], yerr=[2,3], fmt='none', capsize=5, color='red')
# cap size added in line 5
```
