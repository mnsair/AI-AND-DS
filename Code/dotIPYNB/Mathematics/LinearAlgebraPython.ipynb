# Linear Algebra in Python


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Equation
2x + 3y = 8\
5x + 4y = 20

solve for x as below.

```python
# Coefficient matrix
A = np.array([[2,3], [5, 4]])
A
```

```python
# Constant matrix B
B = np.array([8, 20])
B
```

```python
# Solve for x
x = np.linalg.solve(A, B)
x
```

## Solving complex Equations

\[
\begin{align*} 
(1 + 2i)x + (2 - i)y &= 4 + i \\
(3 - i)x + 4y &= 3 + 4i 
\end{align*}
\]


```python
# Coefficients matrix A with cmplex numbers from the complex number set
A = np.array([[2+1j, 3], [5, 4-2j]])

# Constant matrix B with complex numbers from the complex number set
B = np.array([8+2j, 20-4j])

# Solve for x
x = np.linalg.solve(A, B)
x
```

# SVD - singular Value Decomposition

```python
# generate a matrix of 4x3 dataset
A = np.random.rand(4,3)
A
```

```python
# Apply SVD on matrix A
U, Sigma, VT = np.linalg.svd(A)

# Rpint each matrix separately
print("U = ", U)
print("_____________________________________________________")
print("Sigma = ", Sigma)
print("_____________________________________________________")
print("VT = ", VT)
```

```python
# Create a vector
v = np.array([3,3])
v
```

```python
# Create a plot for vactor v
plt.figure()
plt.quiver(0,0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')   
plt.xlim(0,5)
plt.ylim(0,5)    
plt.grid()
plt.title('Vector v')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Linear Transformation on a Vactor

```python
# Apply linear transformation on vector v by 45 degrees
theta = np.radians(55)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], 
                            [np.sin(theta), np.cos(theta)]])

print("Rotation matrix = ", rotation_matrix)
print("_____________________________________________________")
print("Vector v = ", v)

# Applying the transforming
v_transformed = np.dot(rotation_matrix, v)
v_transformed


# plotting the original and transformed vectors
plt.figure()
plt.quiver(0,0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original V')
plt.quiver(0,0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed V')
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.title('Vector v and its transformation')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
plt.figure()
plt.quiver(0,0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original V')
plt.quiver(0,0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed V')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.title('Vector v and its transformation')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
