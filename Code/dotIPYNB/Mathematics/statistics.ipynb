# Inter Quartile Range

```python
# Find IQR for df
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating a dataset with potential outliers
np.random.seed(0)
data = np.random.normal(100, 20, 200)
data = np.append(data, [300, 5]) # Adding potential outliers

#calculating IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
print(iqr)
```

```python
# Defining the outliers criteria (1.5 times the IQR)
upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr

# Identifying the outliers
outliers = data[(data>upper_limit) | (data<lower_limit)]

# Removing the outliers
data_new = data[~np.isin(data, outliers)]

# Plotting the original data with outliers
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(data=data)
plt.title('Original Data with outliers')


# Plotting the data after removng outliers
plt.subplot(1,2,2)
sns.boxplot(data=data_new)
plt.title('Data after removing outliers')

plt.show()
```

```python
# Calculate variance and standard deviation
# Variance is the average of squared deviations from the mean
# Standard deviation is the square root of variance
# Variance is the square of standard deviation
# Variance is used to find the spread of data
# Standard deviation is used to find the spread of data
# Standard deviation is used to find outliers
# Standard deviation is used to find the normal distribution of data
# Standard deviation is used to find the confidence interval
# Confidence interval is used to find the range of values in which a data point is expected to lie with a certain probability

import numpy as np
#create a dataset
df = np.array([2,3,4,5,6])
mean = np.mean(df)
print(mean)
# calculate variance
var = np.var(df, ddof=1) # ddof is degree of freedom is used for sample data, whereas for population data ddof=0
print(var)
standard_deviation = np.sqrt(var)
print(standard_deviation)

```

# Distributions

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating example datasets with different distributions
np.random.seed(0)  # For reproducibility

# Normal distribution data
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Uniform distribution data
uniform_data = np.random.uniform(low=-2, high=2, size=1000)

# Exponential distribution data
exponential_data = np.random.exponential(scale=1, size=1000)

# Plotting the distributions
plt.figure(figsize=(12, 8))

# Normal Distribution Plot
plt.subplot(3, 1, 1)
plt.hist(normal_data, bins=30, color='skyblue', alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Uniform Distribution Plot
plt.subplot(3, 1, 2)
plt.hist(uniform_data, bins=30, color='lightgreen', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Exponential Distribution Plot
plt.subplot(3, 1, 3)
plt.hist(exponential_data, bins=30, color='salmon', alpha=0.7)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Q1: Why normal distribution is important in Data Science or Data Analysis?
The normal distribution is crucial in data science and analysis for several reasons:

Commonality:

Many natural phenomena exhibit a normal distribution, simplifying assumptions and analysis.
Statistical Inference:

Fundamental for statistical tests like t-tests and z-tests, assuming or working best when data follows a normal distribution.
Central Limit Theorem (CLT):

The CLT states that the sum or average of a large number of random variables approaches a normal distribution, aiding in inferential statistics.
Parameter Estimation:

Used in maximum likelihood estimation for model parameter estimation in data science.
Data Transformation:

Normalization or standardization is common in machine learning, improving algorithm performance.
Outlier Detection:

Provides a basis for identifying outliers, aiding in data cleaning and anomaly detection.
Probability Distributions:

Used in probability theory to model uncertainty and calculate probabilities.
Parameter Testing:

Employed in hypothesis testing to assess if sample data differs significantly from a population distribution.
Decision Making:

Guides decision-making processes, especially in quality control and process improvement.

### Q2: How to normalize the data, explain all methods?
#### 1. Min-Max Scaling (Normalization):
   - **Formula:**
     \[ X_{\text{normalized}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}} \]
   - **Explanation:**
     - Scales the data between 0 and 1.
     - Preserves the original distribution but compresses it into the [0, 1] range.

#### 2. Z-Score Normalization (Standardization):
   - **Formula:**
     \[ Z = \frac{X - \mu}{\sigma} \]
   - **Explanation:**
     - Standardizes the data to have a mean (\(\mu\)) of 0 and a standard deviation (\(\sigma\)) of 1.
     - Results in a distribution centered around 0, with values expressed in terms of standard deviations from the mean.

#### 3. Robust Scaling:
   - **Formula:**
     \[ X_{\text{normalized}} = \frac{X - Q_1}{Q_3 - Q_1} \]
   - **Explanation:**
     - Uses the interquartile range (IQR) instead of the min-max range.
     - Robust to outliers, as it focuses on the middle 50% of the data.

#### 4. Log Transformation:
   - **Formula:**
     \[ X_{\text{normalized}} = \log(X) \]
   - **Explanation:**
     - Useful for data with a skewed distribution.
     - Can help stabilize variance across different scales.

#### 5. Softmax Scaling (for Multiclass Classification):
   - **Formula:**
     \[ P(y_i) = \frac{e^{X_i}}{\sum_{j=1}^{n} e^{X_j}} \]
   - **Explanation:**
     - Commonly used in the output layer of a neural network for multiclass classification.
     - Converts raw scores into probabilities, ensuring they sum to 1.

#### 6. Unit Vector Transformation (L2 Normalization):
   - **Formula:**
     \[ X_{\text{normalized}} = \frac{X}{\|X\|_2} \]
   - **Explanation:**
     - Scales each data point so that the overall vector has a length of 1.
     - Useful when the direction of the data points matters more than their magnitudes.

#### 7. Box-Cox Transformation:
   - **Formula:**
     \[ X_{\text{normalized}} = \begin{cases} \frac{X^\lambda - 1}{\lambda} & \text{if } \lambda \neq 0 \\ \log(X) & \text{if } \lambda = 0 \end{cases} \]
   - **Explanation:**
     - Suitable for data that doesn't follow a normal distribution.
     - Adjusts the skewness of the data by using a parameter \(\lambda\).

#### 8. Max Absolute Scaling:
   - **Formula:**
     \[ X_{\text{normalized}} = \frac{X}{\max(|X|)} \]
   - **Explanation:**
     - Scales the data by the maximum absolute value, resulting in a range of [-1, 1].

```python
import numpy as np
from scipy.stats import skew, kurtosis

# Generating example data
np.random.seed(0)
data = np.random.normal(0, 1, 1000)  # Normal distribution

# Calculating skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data, fisher=True)  # 'fisher=True' gives the excess kurtosis

# Displaying the results
print(f"Skewness of the dataset: {data_skewness}")
print(f"Kurtosis of the dataset: {data_kurtosis}")
```

```python
sns.histplot(data = data, kde=True, alpha=0.7)
```
